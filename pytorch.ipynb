```python
import numpy as np
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')
db = client['Dataset']  # Make sure this is the correct database name
collection = db['audio']

# Fetch data for only the first 10 songs
documents = collection.find({}).limit(10)

# Initialize a list to store data
songs_data = []

# Iterate over documents to extract all information
for doc in documents:
    song_info = {
        'object_id': str(doc.get('_id')),
        'filename': doc.get('filename'),
        'metadata': doc.get('metadata'),
        'mfccs_stats': doc.get('features', {}).get('mfccs_stats', {}),
        'spectral_centroid_stats': doc.get('features', {}).get('spectral_centroid_stats', {}),
        'zero_crossing_rate_stats': doc.get('features', {}).get('zero_crossing_rate_stats', {})
    }
    songs_data.append(song_info)

# Display data for the first 10 songs
for song in songs_data:
    print(song)
```
